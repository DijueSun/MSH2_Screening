```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
file_path = '/Users/ds39/Documents/Saturation_genome_editing/MSH2_Screening/2023_April_47470/47470_analysis/E1.1_SGa_count_frame.csv'  # Replace with the path to your file
data = pd.read_csv(file_path)

```

```python
data_D1=data[['D1R1','D3R1','D7R1']]
data_D3=data[['D1R2','D3R2','D7R2']]
data_D7=data[['D1R3','D3R3','D7R3']]
```

```python
import numpy as np
data_D1=np.log(data_D1+0.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns



# Display basic statistics for all the columns
print(data_D1.describe())


# Set up the plot
plt.figure(figsize=(15, 8))

# Plot the distribution of all numeric columns
for column in data_D1.select_dtypes(include=['number']).columns:
    sns.histplot(data_D1[column], kde=False, element="poly", stat="percent", label=column,binwidth=1)

plt.legend(title='Columns')
plt.title('Distribution of All Numeric Columns')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Set up the plot
plt.figure(figsize=(15, 8))

# Plot the distribution of all numeric columns
for column in data_D7.select_dtypes(include=['number']).columns:
    sns.boxplot(data_D7[column])

plt.legend(title='Columns')
plt.title('Distribution of All Numeric Columns')
plt.xlim(0,2000)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
